# 1. Steganography
- Conceal data in another medium. 
    - Payload: Concealed data
    - Carrier: Medium (signal, stream, data file) used to hide the payload
**Image Carriers**
- Image types:
    - Raw: Easiest to work with but very large files and not used by most people (proz and pro s/w) 
    - Lossless compression: Is compressed and data is not modified on opening, transporting or decompressing. 
        - Modify Least-Significant bits (LSB) of carrier to conceal payload
        - Examples: PNG, TIFF, BMP
    - Lossy compression: Is compressed and no guarantee that data will not get modified on storage, transport or decompressing
        - More complex for steganography. 
        - E.G. JPEG, BPG
        
        
**LSB Steganography**
- Steps
    1. Get carrier image details like file format, image size, 
    2. Read payload and add  buffers and any other details required to reconstruct the message. May also encrypt
    3. Determine if carrier is large enough to hide the resulting payload.  Note x8 b/c size in bits not bytes
        - RGB/RGBA images have 3 bits per pizel available while Gray(L) have 1 bit per pixel
        - Carrier capacity be at least x2 of payload 
        
    4. convert payload and buffers into binary data
        - Fill up extra bits (diff payload and carrier capacity) with pseudorandom data
        
    5. If carrier is large enough, alter each pixel LSB in carrier to be corresponding bit from payload. Save as new image
